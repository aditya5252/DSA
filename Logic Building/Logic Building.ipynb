```python
def nearest_num(n,m):
    # If floor function is used 
    div=n//m
    num1, num2= div*m, (div+1)*m
    diff1, diff2 = abs(num1-n), abs(num2-n)
    if(diff1==diff2):
        return num1 if abs(num1)>abs(num2) else num2
    return num2 if diff1>diff2 else num1    
# If int(n/m) is used, then GFG approach
```

```python
def decToBinary(n):
    binArr = []

    while n > 0:
        bit = n % 2
        binArr.append(str(bit))
        n //= 2

    # reverse the string
    binArr.reverse()
    return "".join(binArr)

def bintoDecimal(x):
    string = str(x)[::-1]
    count=0
    for i,s in enumerate(string):
        count+= int(s)*(2**i)
    return count   

decToBinary(1000)
```

```python
def dec_to_bin(x):
    str_ls=[]
    while(x!=0):
        q=x//2
        r=str(x%2)
        str_ls.append(r)
        x=q
    return ''.join(str_ls)[::-1]
    
```

```python
def reverse(text):
    rev_text = ""
    for char in text:
        rev_text = char + rev_text
    return rev_text

def dec_to_bin1(x):
    str_ls=[]
    while(x!=0):
        r=str(x%2)
        str_ls.append(r)
        x//=2
    string_rev = ''.join(str_ls)
    return reverse(string_rev)
    
```
